```python
# Importing important libraries and csv file
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns
%matplotlib inline
df = pd.read_csv('results.csv')
#Checking to see if loaded correctly
df.head()
```

```python
#Seeing the end values of file
df.tail()
```

```python
#Separating out various values in the csv file
x= df['City Average']
y= df['Global Average']
index = df['year']
```

```python
#Plotting the local and global temperature against years and temperature
x_ticks = np.arange(1796, 2013, 30)
plt.xticks(x_ticks, rotation='vertical')
plt.rcParams["figure.figsize"] = (10,5)
plt.plot(index,x,label = "Global temp")
plt.plot(index,y,label = "Local temp")
plt.xlabel('Years')
plt.ylabel('Temperature')
plt.title('Comparing local and global temperature ')
plt.legend()
plt.show()
```

```python
#Plotting the moving average of local and global temperature against years and temperature

x_ticks = np.arange(1796, 2013, 10)
plt.xticks(x_ticks,rotation='vertical')
x_mva = x.rolling(10).mean()
y_mva = y.rolling(10).mean()
plt.plot(index, x_mva,label = "Global temp")
plt.plot(index,y_mva,label = "Local temp")
plt.xlabel('Years')
plt.ylabel('Temperature')
plt.title('Comparing moving averages of local and global temperature ')
plt.legend()
plt.show()
```
